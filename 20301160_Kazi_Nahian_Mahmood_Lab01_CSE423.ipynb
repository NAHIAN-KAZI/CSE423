```python
import numpy as np

import ctypes.util
from lucid.misc.gl.glcontext import create_opengl_context

# Now it's safe to import OpenGL and EGL functions
import OpenGL.GL as gl
from OpenGL.GLU import *

# create_opengl_context() creates GL context that is attached to an
# offscreen surface of specified size. Note that rendering to buffers
# of different size and format is still possible with OpenGL Framebuffers.
#
# Users are expected to directly use EGL calls in case more advanced
# context management is required.
WIDTH, HEIGHT = 1200,900
create_opengl_context((WIDTH, HEIGHT))
    
# OpenGL context is available here.

print(gl.glGetString(gl.GL_VERSION))
print(gl.glGetString(gl.GL_VENDOR)) 
#print(gl.glGetString(gl.GL_EXTENSIONS))
```

```python
def  addvertex(x,y):
  gl.glVertex2f(x/(WIDTH/2),y/(HEIGHT/2))
```

```python
from IPython.display import display
from PIL import Image
import random
# Let's render something!
gl.glClear(gl.GL_COLOR_BUFFER_BIT)
gl.glPointSize(6)
gl.glBegin(gl.GL_POINTS)

for i in range(50):
  x = random.randint(-599,499)
  y = random.randint(-499,599)
  addvertex(x,y)  
  r = random.uniform(0,1.0)
  g = random.uniform(0,1.0)
  b = random.uniform(0,1.0)
  gl.glColor3f(r,g,b)

gl.glEnd()

img_buf = gl.glReadPixelsub(0, 0, WIDTH, HEIGHT, gl.GL_RGB, gl.GL_UNSIGNED_BYTE)
img = np.frombuffer(img_buf,np.uint8).reshape(HEIGHT, WIDTH, 3)[::-1]
display(Image.fromarray(img,'RGB'))
```

```python
from IPython.display import display
from PIL import Image
import random
# Let's render something!

gl.glClear(gl.GL_COLOR_BUFFER_BIT)
#AXIS
gl.glColor3f(0,10,90)
gl.glBegin(gl.GL_LINES)
gl.glVertex2f(0,1)
gl.glVertex2f(0,-1)
gl.glVertex2f(1,0)
gl.glVertex2f(-1,0)
gl.glEnd()
gl.glLineWidth(5)

def zero(a):
  gl.glColor3f(0,100,0)
  gl.glBegin(gl.GL_LINES)
  addvertex(a,120)
  addvertex(a,-120)

  addvertex(a,120)
  addvertex(a+120,120)

  addvertex(a+120,120)
  addvertex(a+120,-120)

  addvertex(a,-120)
  addvertex(a+120,-120)
  gl.glEnd()

def three(a):
  gl.glColor3f(1,1,1)
  gl.glBegin(gl.GL_LINES)
  addvertex(a,-120)
  addvertex(a+120,-120)

  addvertex(a,0)
  addvertex(a+120,0)

  addvertex(a,120)
  addvertex(a+120,120)

  addvertex(a+120,-120) 
  addvertex(a+120,120)
  gl.glEnd()

def one(a):
  gl.glColor3f(1,1,0.5)
  gl.glBegin(gl.GL_LINES)
  addvertex(a,120)
  addvertex(a,-120)
  gl.glEnd()

def two(a):
  gl.glColor(2,0.5,2)
  gl.glBegin(gl.GL_LINES)
  addvertex(a,0)
  addvertex(a+120,0)

  addvertex(a+120,120)
  addvertex(a,120)

  addvertex(a+120,120)
  addvertex(a+120,0)

  addvertex(a,0)
  addvertex(a,-120)

  addvertex(a,-120)
  addvertex(a+120,-120)

  
  gl.glEnd()

def six(a):
  gl.glColor(100,100,0)
  gl.glBegin(gl.GL_LINES)
  addvertex(a,120)
  addvertex(a,-120)

  addvertex(a,-120)
  addvertex(a+120,-120)

  addvertex(a+120,-120)
  addvertex(a+120,0)

  addvertex(a+120,0)
  addvertex(a,0)

  addvertex(a,120)
  addvertex(a+120,120)

  gl.glEnd()
two(-580) 
zero(-440)
three(-300)
zero(-150)
one(50)
one(150)
six(250)
zero(450)
img_buf = gl.glReadPixelsub(0, 0, WIDTH, HEIGHT, gl.GL_RGB, gl.GL_UNSIGNED_BYTE)
img = np.frombuffer(img_buf,np.uint8).reshape(HEIGHT, WIDTH, 3)[::-1]
display(Image.fromarray(img,'RGB'))
```
